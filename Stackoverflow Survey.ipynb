```python
#increasing size of the plot
fig, plot = plt.subplots(figsize = (20,10))
#visualizing age counts of the participants
numberOfParticipants = sns.countplot(data=df, x="Age")
plt.title("Participants")
plt.ylabel("Participant Count", size = 20)
plt.xlabel("Age", size = 20)
```

```python
fig, scatter = plt.subplots(figsize = (40,20))
#visualizing gender counts of participants
numberOfParticipants = sns.countplot(data=df, x="Gender")
#defining yaxis limit to see the gender count better
plt.ylim(0, 40000)
#seeing only two genders man and woman in this case
plt.xlabel("Gender",size=30)
plt.ylabel("Count",size=30)
plt.xlim(-0.5,1.5)
```

```python
#visualizing education level counts 
fig, plot = plt.subplots(figsize = (60,30))
#visualizing gender counts of participants
sns.countplot(data=df, x="EdLevel")
plt.xlabel("Education Level",size=50)
plt.ylabel("Count",size=50)
```

```python
sns.relplot(data=df, x=df[adultsInCanada]["Age"], y=df[adultsInCanada]["ConvertedComp"], kind="line")
plt.xlabel("Age", size=20)
plt.ylabel("Salary", size = 20)
plt.title("Yearly Salary in Canada By Adults USD$ ")
#ignoring outliar
plt.xlim(20,)
```

```python
fig, plot = plt.subplots(figsize = (80,50))
filt = df["CompFreq"].str.contains("Yearly") == True
sns.barplot(data=df[filt]["ConvertedComp"], x=df[filt]["Age"], y=df[filt]["ConvertedComp"])
```

```python
#filter to get only yearly paid rows
filt =  df["CompFreq"].str.contains("Yearly") == True
```

```python
#creating a new row for yearly salary
df["Yearly"] = df[filt]["ConvertedComp"]
```

```python
#filling na values with mean value so we keep the mean same
df["Yearly"].fillna(df["Yearly"].mean(), inplace=True)
```

```python
#visualizing the yearly salary based on age
fig, plot = plt.subplots(figsize = (80,50))
filt = df["Age"] > 18
sns.barplot(data=df, x=df[filt]["Age"], y=df["Yearly"])
```
